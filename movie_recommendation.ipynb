```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        
        else:
                break

    return L






```

```python
 movies['cast']=movies['cast'].apply(convert3)
```

```python
import ast

def fetch_director(obj):
    try:
        # Safely evaluate the string to a Python object
        crew_list = ast.literal_eval(obj)
        
        # Ensure the evaluated object is a list
        if isinstance(crew_list, list):
            for i in crew_list:
                if i.get('job') == 'Editor':  # Check if 'job' exists and matches 'Editor'
                    return i.get('name', '')  # Return the first match
    except (ValueError, SyntaxError):
        pass  # Handle invalid or malformed input gracefully
    
    return ''  # Return an empty string if no 'Editor' is found
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
import ast

def clean_spaces_in_list(obj):
    # Ensure obj is a list
    try:
        obj = ast.literal_eval(obj) if isinstance(obj, str) and obj.startswith('[') else obj
    except (ValueError, SyntaxError):
        return obj  # If obj cannot be parsed, return as is.

    if isinstance(obj, list):
        cleaned = []
        for i in obj:
            if isinstance(i, dict):  # If it's a dictionary, clean keys and values
                cleaned.append({k.replace(" ", ""): v.replace(" ", "") if isinstance(v, str) else v for k, v in i.items()})
            elif isinstance(i, str):  # If it's a string, just clean spaces
                cleaned.append(i.replace(" ", ""))
            else:
                cleaned.append(i)  # Leave other data types unchanged
        return cleaned
    return obj
    
movies['genres'] = movies['genres'].apply(clean_spaces_in_list)
movies['keywords'] = movies['keywords'].apply(clean_spaces_in_list)
movies['cast'] = movies['cast'].apply(clean_spaces_in_list)
movies['crew'] = movies['crew'].apply(clean_spaces_in_list)
```

```python
movies['tags'] = (
    movies['overview'].astype(str) +
    " " +
    movies['genres'].astype(str) +
    " " +
    movies['keywords'].astype(str) +
    " " +
    movies['cast'].astype(str) +
    " " +
    movies['crew'].astype(str)
)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):

    y = []
    for i in text.split():
        y.append(ps.stem(i))
        return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
vectors = cv.fit_transform(new_df['tags']).toarray()

```

```python
# Get feature names using the new method
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title) 
```

```python
recommend('Batman Begins')
```
